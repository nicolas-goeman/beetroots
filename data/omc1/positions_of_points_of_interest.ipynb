values from "Molecular tracers of radiative feedback in Orion (OMC-1)", Goicoechea et al. 019

```python
import pandas as pd
import numpy as np
```

```python
data = np.array(
    [
        [1, 76.0, 305.0],
        [2, -25, 178],
        [3, -38, -50],
        [4, 51, -165],
        [5, -25, -291],
        [6, 51, 166],
        [7, 114, 13],
        [8, 76, -63],
        [10, -114, 1],
    ]
)

df = pd.DataFrame(data, columns=["number", "x", "y"])
df
```

```python
def from_offset_to_pos_theta(x):
    return round(20 - 20 * x / 235.143)


def from_offset_to_pos_y(x):
    return round(30 + 30 * x / 350)
```

```python
df["x"] = df["x"].apply(from_offset_to_pos_theta)
df["y"] = df["y"].apply(from_offset_to_pos_y)
```
